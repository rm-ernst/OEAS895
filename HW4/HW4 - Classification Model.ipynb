# MLP Classification Model
## Part One: Model Construction
Construct a model using the MLPClassifier to predict the target variable for the dataset: 'wine'

UCI ML Wine recognition dataset via sklearn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.neural_network import MLPClassifier

wine = load_wine()
print(wine.target)
```

```python
nrows, ncols = 4, 3
fig = plt.figure(figsize = (7,8), dpi = 300)

for i in range(1, 13):
    ax = fig.add_subplot(nrows, ncols, i)
    ax.scatter(wine.data[:,i-1], wine.data[:,i],marker=".")
    ax.set_xlabel(wine.feature_names[i-1])
    ax.set_ylabel(wine.feature_names[i])

plt.tight_layout()
plt.show()
```

A quick plot of various features doesn't show anything notable / outlying in the dataset, nor give an immediate impression of how many groups there are. However, the dataset has three targets (class_0, class_1, and class_2.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(wine.data,wine.target,test_size=0.2,random_state=2)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Part Two: Explore hyperparameter space

To explore the hyperparameter space, a maximum number of layers and neurons was set at 20 each.

```python
layer = []
node = []
acc = []
from sklearn import metrics

for i in range(1,21):
    for j in range(1,21):
        t = np.tile(i,j)
        model = MLPClassifier(hidden_layer_sizes=(t),random_state=1)
        model.fit(x_train_scaled,y_train)
        y_pred = model.predict(x_test_scaled)
        node.append(i)
        layer.append(j)
        acc.append(metrics.accuracy_score(y_test,y_pred))
dict = {'layer':layer,'node':node,"accuracy":acc}
df = pd.DataFrame(dict)
df
```

```python
plt.scatter(df.layer,df.node,marker="o",c=df.accuracy, cmap='magma')
plt.xlabel('number of layers')
plt.ylabel('number of nodes')
plt.colorbar()
```

In general, accuracy increased with a higher number of nodes, but decreased slightly with number of layers, more notably on the higher end - with layers=20 having low values even with high number of nodes. 

There are 200+ combinations >0.9 so I will not be doing confusion matrixes for all of them. Instead, I have narrowed it down by the following reasons - 
- less than 1, as a 1.0 match is certainly overfit
- nodes less than 15, to avoid overfit
- layers less than 15, as layers decrease in accuracy as number increased.

```python
newdf = df[df.accuracy > 0.9]
newdf = newdf[newdf.accuracy < 1]
newdf = newdf[newdf.node < 15]
newdf = newdf[newdf.layer < 15]
newdf
```

```python
lng = newdf.layer.count()
from sklearn.metrics import classification_report, confusion_matrix

for i in range(0,lng):
    node = newdf.iloc[i]['node']
    node = np.int64(node)
    layer = newdf.iloc[i]['layer']
    layer = np.int64(layer)
    t = np.tile(node,layer)
    model = MLPClassifier(hidden_layer_sizes=(t),random_state=1)
    model.fit(x_train_scaled,y_train)
    y_pred = model.predict(x_test_scaled)
    print("layer: ",layer," nodes: ",node)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
```

The majority of F1 values were greater than 0.9 and of those less than 0.9
- 0.84 in class 1 for 3,5; 7,7; 3,11; 1,12; 3,12
- 0.86 in class 1 for 2,7; 2,8; 3,8; 10,9; and class 2 for 12,13
- 0.88 in class 1 for 4,6 and 2,13 and class 2 for 13,10
- 0.89 in class 1 for 2,5; 1,6; 3,7; 14,8; 3,9; 5,12; 8,13; 11,13; and class 2 for 12,8 and 3,9
which generally shows a trend of class 1 being harder to predict.

For overall accuracy, the highest value was 0.97 which several combinations reached for macro avg and weighted avg as well. (4,5; 5,5; 5,8; 9,8; 2,10; 6,10; 5,11; 7,11; 12,11; 8,12; 10,13; 13,13; 10,14; 11,14)

All had confusion matrixes with only one error (as 1.0 values were excluded), with most having a class 0 as a class 1, but a few miscategorizing a class 1 or class 2 as class 0 instead, as seen below.

| Majority  |   |   | 6,10; 8,12  |   |   |10,13   |   |   |

|17 | 1 | 0 |18 | 0 | 0 |18 | 0 | 0 |

As these had the same accuracy and were generally identical, the lowest number of nodes and layers was chosen - layers=4 and nodes=5
